```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data
data = pd.read_csv("podatki_imena.csv")
```

```python
# Replace Nan values
data = data.replace({'-': '0'}, regex=True)
```

```python
#Creating headers that will be used later
header = list(data.columns.values)
years = header[1:]
sort_by_list = [False for i in range(len(years))]
```

```python
# Change all number in string to int
data[years] = data[years].apply(pd.to_numeric)
#Sort data
sorted_data = data.sort_values(by=years[::-1], ascending=sort_by_list)
sorted_data
```

```python
#Ploting
N = 10
width = 0.5
y_pos = np.arange(N)

plot_value = list(sorted_data['2016'][:N].values)
#print (plot_value)
plot_name = list(sorted_data['Ime'][:N].values)
#print (plot_name)

plt.bar(y_pos, plot_value, width, color="blue")
plt.xticks(y_pos, plot_name)
plt.ylabel('Pojavitve imena')
plt.title('Moska imena v letu 2016 v Sloveniji')


plt.show()
```

```python
# Now I will try to get difference betwen two years
year_2008 = data['2008'].values
year_2016 = data['2016'].values
```

```python
# Calculating difference 
diff = []

for i in range(len(year_2008)):
	diff.append(year_2016[i] - year_2008[i])
```

```python
# Putting together data
diff_data = pd.DataFrame(data={'Ime': data['Ime'], 'Razlika': diff}, index=None, columns=None, dtype=None, copy=False)
diff_data
```

```python
# Sorting difference
sorted_diff_data = diff_data.sort_values(by="Razlika", ascending=False)
sorted_diff_data
```

```python
# Plotting
plot_value = list(sorted_diff_data['Razlika'][:N].values)
plot_name = list(sorted_diff_data['Ime'][:N].values)

plt.bar(y_pos, plot_value, width, color="blue")
plt.xticks(y_pos, plot_name)
plt.ylabel('Sprememba v stevilu imen')
plt.title('Sprememba moskih imen med letoma 2008 in 2016 v Sloveniji')

plt.show()
```

```python
# Plotting
plot_value = list(sorted_diff_data['Razlika'][(-N):].values)
plot_name = list(sorted_diff_data['Ime'][(-N):].values)

plt.bar(y_pos, plot_value, width, color="red")
plt.xticks(y_pos, plot_name)
plt.ylabel('Sprememba v stevilu imen')
plt.title('Sprememba moskih imen med letoma 2008 in 2016 v Sloveniji')

plt.show()
```
